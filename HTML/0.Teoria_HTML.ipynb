**Estructuras HTML**

- *Headingd(encabezados)* : Van desde el h1 hasta el h6, cada uno de ellos tiene un estilo y tamao propio; los headings por defecto no tienen ningun tipo de sangria ni de espaciado, para plasmar en un heading o en cualquier otra etiqueta de escritura espaciados o sangrias se deben hacer con etiquetas especiales



- *Paragraphs (parrafos)* : Son cadenas de texto grandes las cuales se pueden escribir con la etiqueta <p>

- *Format (Formatos)* : Se refiere al formato de letra que van a tener todos los caracteres que van a estar presentes en la pagina web ; estos formatos pueden ser :



- *Atributes :* son utiles para agregar funcionalidad a la pagina, por ejemplo el atributo 'href = ' nos sirve para enlazar una pagina externa con un link; tambien, dentro de la etiqueta img aparecen 2 atributos : 

    - 'src' : indica la ruta donde esta la imagen
    - 'alt' : indica un texto que puede ser de error cuando la imagen no esta disponible 

- *Lists :* pueden ser de 2 tipos, ordenadas o desordenadas; para hacer listas en html se hace de la siguiente manera : 
    - Listas ordenadas : ol seguido de li
    - Listas desordenadas : ul seguido de li
